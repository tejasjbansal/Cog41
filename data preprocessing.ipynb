```python
# Import the Data preprocessing Libraries

import numpy as np
import pandas as pd
```

```python
# Read the Customers data from the csv file

df = pd.read_csv("CustomersData/BankCustomerData.csv")
df.head()
```

```python
# use only top 10 rows of the data

customers_details = df.head(10)
```

```python
# Drop the useless columns

customers_details = customers_details.drop(['default','balance','housing','loan','contact','day','campaign','pdays','previous','poutcome','term_deposit'],axis=1)
```

```python
# After droping useless columns and rows

customers_details
```

```python
# Reading another customer transaction dataset

df2 = pd.read_csv("CustomersData/accountstatements.csv")
df2.head()
```

```python
# Print the columns of the above dataset

df2.columns
```

```python
# Clean the Account No column by removing unwanted character

df2['Account No'] = df2['Account No'].apply(lambda val: val[0:-1])
```

```python
# Drop useless column

df2.drop(['.'],axis=1,inplace=True)
```

```python
# clean the dataset by replacing nan values by ----

def replace_nan(col_list):
    for col in col_list:
        df2[col] = df2[col].replace(np.nan,'---')

replace_nan(df2.columns)
```

```python
# Save the dataset to the new csv file

df2.to_csv("CustomersData/newaccountstatements.csv")
```

```python
# creating unique accounts numbers array for customers_details dataset

Account_numbers = df2["Account No"].unique()
Account_numbers
```

```python
#  Add the new Account No column in customers_details dataset and also modify some month column

customers_details["Account No"] = Account_numbers
customers_details['month'] = ["apr","feb","mar","apr","dec","nov","jun","jul","aug","dec"]
customers_details
```

```python
#  Save the dataset to the new csv file

customers_details.to_csv("CustomersData/customers_details.csv")
```
